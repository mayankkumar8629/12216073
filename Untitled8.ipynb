```python
import yfinance as yf
ticker_symbol = "MSFT"
stock_data = yf.download(ticker_symbol,start='2024-01-01',end='2025-10-17')
print(stock_data)
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from datetime import datetime, timedelta

# 1. Use only the 'Close' price
data = stock_data[['Close']].copy()
data = data.dropna()

# 2. Scale data between 0 and 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data)

# 3. Convert to time sequences (last 60 days → predict next day)
look_back = 60
X_train, y_train = [], []
for i in range(look_back, len(scaled_data)):
    X_train.append(scaled_data[i-look_back:i, 0])
    y_train.append(scaled_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))  # [samples, timesteps, features]

# 4. Simple LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

print("Training model... please wait")
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# 5. Predict future until Oct 26, 2025
last_date = data.index[-1]
future_end = datetime(2025, 10, 26)

x_input = scaled_data[-look_back:].reshape(1, look_back, 1)
future_preds = []
future_dates = []

while last_date < future_end:
    pred = model.predict(x_input, verbose=0)[0,0]
    future_preds.append(pred)

    # shift window forward by one
    x_input = np.append(x_input[:, 1:, :], [[[pred]]], axis=1)
    last_date += timedelta(days=1)
    future_dates.append(last_date)

# 6. Inverse scale predictions
future_preds = scaler.inverse_transform(np.array(future_preds).reshape(-1, 1))

# 7. Plot
plt.figure(figsize=(12,5))
plt.plot(data.index, data['Close'], label='Historical Close')
plt.plot(future_dates, future_preds, linestyle='--', label='Predicted Close')
plt.title("MSFT Stock Price Prediction Until 2025-10-26")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()

# 8. Show forecasted values
forecast_df = pd.DataFrame({'Date': future_dates, 'Predicted_Close': future_preds.flatten()})
print("\nPredicted Prices:")
print(forecast_df.tail(10))
```
